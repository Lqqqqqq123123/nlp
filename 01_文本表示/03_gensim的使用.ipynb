```python
from gensim.models import KeyedVectors
from requests import post

model_path = r'./data/sgns.weibo.word.bz2'
model = KeyedVectors.load_word2vec_format(model_path)
```

```python
print(len(model.index_to_key))
print(len(model.key_to_index))
```

```python
print(model.vectors[0])
```

```python
print(model.similarity('男孩', '男人'))
```

```python
model.similar_by_word('男孩')
```

```python
model.most_similar(positive=['中国'])
```

```python
model.most_similar(positive=['男人', '女孩'], negative=['女人'], topn=4)
```

```python
# 2. 自行训练 wordtovec 模型
from gensim.models import Word2Vec

net = Word2Vec(
    sentences=[
        ['我', '每天', '乘坐', '地铁', '上班'],
        ['我', '每天', '乘坐', '公交', '上班']
    ],
    vector_size=32,  # 词向量维度
    window=1,  # 上下文窗口大小
    min_count=0,  # 最小词频（低于此值的词会被忽略）
    sg=1,  # 1: Skip-Gram，0: CBOW
    workers=4  # 并行训练线程数
)
```

```python
print(net.vector_size)
print(net)
```

```python
net.wv.vector_size
```

```python
net.wv.vectors
```

```python
net.wv.similarity('地铁', '公交')
```
