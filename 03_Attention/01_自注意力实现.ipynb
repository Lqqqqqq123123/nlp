```
from math import sqrt

import torch
import torch.nn as nn
torch.set_printoptions(precision=4, sci_mode=False)
```

```
d_q, d_k, d_v = 64, 64, 64
l = 4 # 序列长度
```

```
Q, K, V = torch.randn((l, d_q)), torch.randn((l, d_k)), torch.randn((l, d_v))
```

```
# 1. 计算注意力分数 (n * n)
attn_scores = Q @ K.T
print(attn_scores)
```

```
# 2. 经过 softmax 得到注意力权重
attn_weights = torch.softmax(attn_scores, dim=-1)
print(attn_weights)
```

```
# 3. 注意力权重和 v 加权求和，得到增强后的表示
attn_output = attn_weights @ V
```

```
print( attn_output)
```
